```python
# imports
import numpy as np
import pandas as pd

from keras.models import Sequential
from keras.layers import Dense

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
train_data = pd.read_csv("C:\\DEV\\study\\superdatascience\\Machine Learning A-Z Template Folder\\train.csv")
```

```python
avg_age = train_data["Age"].mean()
std_age = train_data["Age"].std()
cnt_nan_age = train_data["Age"].isnull().sum()

random_ages = np.random.randint(avg_age-std_age, avg_age+std_age, size=cnt_nan_age)
train_data.loc[np.isnan(train_data["Age"]), "Age"] = random_ages
```

```python
y = train_data.iloc[:, 1].values
train_data = train_data.drop('Name', axis=1)
train_data = train_data.drop('Survived', axis=1)
train_data = train_data.drop('Cabin', axis=1)
train_data = train_data.drop('Embarked', axis=1)
train_data = train_data.drop('Ticket', axis=1)
```

```python
x = train_data.iloc[:, 1:].values
```

```python
labelencoder_x_gender = LabelEncoder()
x[:, 1] = labelencoder_x_gender.fit_transform(x[:, 1])
```

```python

onehotencoder = OneHotEncoder(categorical_features=[0])
x = onehotencoder.fit_transform(x).toarray()
x = x[:, 1:]
```

```python
# Feature scaling

sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
# Splitting dataset into Training and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Sequential model is a linear stack of layers

classifier = Sequential()

# This is a common answer, but experiments can be done to find a better node number
# hidden layer nodes number = avg(number of nodes input and output layers)
# (11 + 1)/ (1+1) = 6
# adding first hidden layer
classifier.add(Dense(units=4, kernel_initializer='uniform', activation='relu', input_dim=7))

# adding second hidden layer
classifier.add(Dense(units=4, kernel_initializer='uniform', activation='relu'))

# adding the output layer
# softmax function if you have more than 2 categories
classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
# Compiling the ANN
classifier.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
```

```python
# Fitting the ANN to the training set
classifier.fit(x_train, y_train, batch_size=1, epochs=500)
```

```python
y_pred = classifier.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```
